```python
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/indianspice/Visualizations/master/fifa/fifa.csv'
data = pd.read_csv(url, index_col = "Date", parse_dates=True)

data.head()
```

### Plot the data

```python
plt.figure(figsize=(16, 6))

sns.lineplot(data=data).set_title("FIFA Rankings by Country")
```
